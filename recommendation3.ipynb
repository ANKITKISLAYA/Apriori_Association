# here i have imported aprioi and association rules

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_excel('orderdetails.xlsx')
```

 # this is data from the database on orderid and itemid

```python
df.head()  
```

```python
df.count()
```

# here i grouped my data on orderid  to get desired format


```python
a=df.groupby('OrderId')['ItemId'].apply(list)
```

```python
b=a.to_frame()
```

```python
b=b.reset_index()
```

```python
c=b.drop('OrderId',axis=1)
```

```python
c.head()
```

```python
 c=c.head(1000)
```

```python
dataset=list(c['ItemId'])
```

# this is the data which i have prepared in transaction format for further processing

```python
dataset[1:25]  
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df1 = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df1.head()
```

```python
df2=df1.astype("int")
```

```python
df2.head()
```

# here i am using apriori algorithm 

```python
frequent_item = apriori(df2, min_support=0.005, use_colnames=True)
```

```python
frequent_item.head()
```

```python
frequent_item.tail()
```

# Here i applied association rules

```python
rules = association_rules(frequent_item,metric = "lift" ,min_threshold=1)
```

```python
rules.head()
```

```python
rules.tail()
```

# here i applied filtering on rules is this necessary if yes then what should be ideal values for lift or confidence

# and for the sake of deployment what should be approach like if they give itemid we will search in antecedents and give  them consequents is this correct ?

```python
rules[ (rules['lift'] >= 3) & (rules['confidence'] >= 0.05)]
```

```python
rules[ (rules['lift'] >= 3) & (rules['confidence'] >= 0.05)].tail()
```
